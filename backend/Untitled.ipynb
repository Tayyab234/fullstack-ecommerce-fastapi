```python
from pymongo import MongoClient
import json

# Connect to MongoDB
client = MongoClient("mongodb://localhost:27017/")

# Create / Access Database
db = client["ecommerce"]

# Create / Access Collection
products_collection = db["products"]

# Read JSON file
with open("products.json", "r", encoding="utf-8") as file:
    products = json.load(file)

# Insert products into MongoDB
if isinstance(products, list):
    products_collection.insert_many(products)
    print(f"{len(products)} products inserted successfully.")
else:
    products_collection.insert_one(products)
    print("1 product inserted successfully.")

print("Database: ecommerce")
print("Collection: products")
```
